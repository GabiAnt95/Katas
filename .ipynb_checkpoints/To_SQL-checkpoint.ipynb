Para leer los archivos

```python
customers = pd.read_csv("data/Todos/customers_limpio.csv")
employees = pd.read_csv("data/Todos/employees_limpio.csv")
offices = pd.read_csv("data/Todos/offices.csv")
orders = pd.read_csv("data/Todos/orders_clean.csv")
orders_details = pd.read_csv("data/Todos/orderdetails.csv")
payments = pd.read_csv("data/Todos/payments_clean.csv")
product_lines = pd.read_csv("data/Todos/productlines.csv")
products = pd.read_csv("data/Todos/products.csv")
```

Para hacer la conexión, 'lenguaje+driver://user:contraseña@direccionBD/Base_de_datos'

y el motor

```python
str_conn='mysql+pymysql://root:Password@localhost:3306/Trabajo_grupal'

motor=create_engine(str_conn)
```

Se lleva con 'to_sql' a la Base de datos sin índices

```python
customers.to_sql(name='customers', con=motor, if_exists='append', index=False) # motor con sqlalchemy
employees.to_sql(name='employees', con=motor, if_exists='append', index=False) # motor con sqlalchemy
offices.to_sql(name='offices', con=motor, if_exists='append', index=False) # motor con sqlalchemy
orders.to_sql(name='orders', con=motor, if_exists='append', index=False) # motor con sqlalchemy
orders_details.to_sql(name='orders_details', con=motor, if_exists='append', index=False) # motor con sqlalchemy
payments.to_sql(name='payments', con=motor, if_exists='append', index=False) # motor con sqlalchemy
product_lines.to_sql(name='product_lines', con=motor, if_exists='append', index=False) # motor con sqlalchemy
products.to_sql(name='products', con=motor, if_exists='append', index=False) # motor con sqlalchemy
```
