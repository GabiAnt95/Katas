```python
def decompose(n):

'''Se le da el número a descomponer.
Devuelve una lista formada por los números cuyas sumas al cuadrado son iguales a n al cuadrado.
Si hay varias soluciones, da la solución mayor.
'''
''' Número negativo da None, y 0 da una lista vacía'''
    if n<0: 
        return None
    if n==0:
        return []
    
'''Se define la lista, formada por todos los cuadrados menores de n en orden descendente'''
    l_sq = [i**2 for i in range(n+1)][::-1]

''' Se define una función recursiva'''

    def loop(sq, x):
    '''Se le da sq, que es n**2 y x que es n (va cambiando con cada loop).
    A no ser que se resuelva el ejericicio, devuelve la ejecución de sí misma.
    '''
        
    '''En cada bucle borra el primer elemento de la lista, ya que si no se ha resuelto para ese número,
    significa que no hay solución en la que él esté presente'''
    
        l_sq.pop(0)
        l_fin = []

    ''' Si se llega a un punto en el que la suma de todos los elementos de la lista es menor que n**2,
    se devuelve None, ya que no hay solución'''
        if sum(l_sq) < sq:
            return None
        
    '''Bucle en el que si el número de la lista en el que se está iterando es menor que n**2, a n**2 se le resta
    ese número y se añade a la lista, si se llega a tener n**2 == 0 restándole los números que sean, 
    tenemos una solución y se devuelve dicha lista'''
        for i in l_sq:

            if sq>=i:
                sq -= i
                l_fin.append(i)

                if sq == 0:
                    return [int(i**(1/2)) for i in l_fin][::-1]
                
        return loop(n**2, n)
    
    return loop(n**2, n)
```
