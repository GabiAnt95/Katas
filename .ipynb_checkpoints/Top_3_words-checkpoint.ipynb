Saca el top 3 de palabras en un texto

```python
def top_3_words(text):
    
    from collections import Counter
    
    text = text.lower()
    
    list_text = text.split()
    
    list_text_new = []

    '''Se hace una lista con cada una de las palabras del texto, quitando comas, sÃ­mbolos, puntos, etc'''
    for i in list_text:
        pal = ""
        for j in i:
            if j.isalpha() == False and j != "'":
                
                continue
            else:
                pal = pal + j 

        list_text_new.append(pal)

    list_text_def = [i for i in list_text_new if len(i)>0]
    
    l_def = [i for i in list_text_def if not set(i) == {"'"}]

    if len(l_def) < 3:
        top = Counter(l_def).most_common()[0:len(l_def)]
        
        res = []
        for i in top:
            res.append(i[0])
        return res
                
    top = Counter(l_def).most_common()[0:3]
    
    res = []
    
    for i in top:
        res.append(i[0])
        
    return res.sort(key=lambda x: x[1])
    
        
    
    
    
```
