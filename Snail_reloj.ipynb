```python
def snail(array):
    '''Le das una matriz cuadrada.
    Devuelve una lista de números resultado de recorrer en el sentido de las agujas del reloj y hacia adentro
    dicha matriz (emepzando por el elemento[1,1]).
    '''
    
    import numpy as np
    nparr = np.array(array)
    dim = len(array[0])
    res = []
    cols = []
    c = []
    
    if dim == 0:
        return []
    
''' Se define dentro de cols la misma matriz pero transponiendo cada fila por separado, es decir, se tiene
un array en el que cada lista es una columna, mientras que dentro de "nparr" tenemos un array en el que
cada lista es una fila de la matriz'''

    for i in range(0,dim):
        cols.append(nparr.T[i])
        
''' Se pasa a lista '''

    for i in cols:
        c.append(i.tolist())
    
''' Se hace la parte central del código, se va a añadiendo a la lista "resultado" los números de cuatro
en cuatro. 1): el primer array de las filas exceptuando el ulti elemento.
           2) el ult array de las columnas excep el ult elemento
           3) ult array de las filas excep el ult elem pero dado la vuelta
           4) primer array de las columnas excepo el ult elemen todo dado la vuelta
           5), 6), etc --> Lo mismo pero quitando los elementos de los extremos'''
    for i in range(0,dim):
        res.append(array[i][i:(dim-1-i)])
        res.append(c[dim-1-i][i:(dim-i)-1])
        res.append(array[dim-i-1][1+i:(dim-i)][::-1])
        res.append(c[i][i+1:dim-i][::-1])

'''Se le añade el elemento del medio por si la dim es impar'''
    res.append([array[int(dim/2)][int(dim/2)]])
    
    fin = []
''' Se transofrma en una única lista con todos los números'''    
    for i in res:
        for j in i:
            if j != "":
                fin.append(j)

'''Si es impar se deja como está, si es par no se añade el último número, que se´ria el de el medio si fuese impar'''
    if dim % 2 != 0:
        return fin
    else:
        return fin[:-1]
    
    
    
    
```
