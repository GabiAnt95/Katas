```python
' Función recursiva'
def get_last_digit(index):
    
    fib = []
    def rec_fib(index):
        if index > 1:
            return rec_fib(index - 1) + rec_fib(index-2)
        return index
    
    for i in range(index):
        fib.append(rec_fib(index))
        
    return fib
```

```python
def last_fib_digit(n):
    '''
    Bucle while en el que va añadiendo a fib cada número más el anterior. 
    Función muy resumida.
    '''
    fib = [0,1]
    a, b, cont = 0, 1, 1
    while cont <= n:
        a, b, cont = b, (a+b), + cont + 1
        fib.append(a)
        
    return fib
```

```python
'Bucle while'

def get_last_digit(index):
    ''' Se le da el índice al que se quiere llegar.
    Devuelve la serie.
    
    Va añadiendo a la lista la suma más el número anterior, siendo "suma", n y "ant", n-1.
    '''
    
    ant = 0
    fib = [0, 1]
    suma = 1
    
    while len(fib) <= index:
        fib.append(suma + ant)
        ant = fib[-2]
        suma = fib[-1]
        
    return fib
```
